```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("titanic-passengers.csv",encoding="iso.8859-1", sep=";")
df.head()
```

```python
correlation = df['Sex'].str.get_dummies(sep=' ').corrwith(df['Age'])
```

```python
plt.ylabel('Correlation between  Age & Sex')
plt.plot(correlation)
plt.show()
```

```python
sns.barplot(data=df, x='Pclass', y='Fare', hue ='Survived')
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df)
print("There's no Heatmap correlation. The heatmap allows us to only have a correlation matrix to visualize at once the state of our data")
```
